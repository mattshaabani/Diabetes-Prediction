```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
dataSet_diabetes = pd.read_csv('diabetes.csv')
```

```python
dataSet_diabetes.head()
```

```python
# @title SkinThickness vs Insulin

from matplotlib import pyplot as plt
dataSet_diabetes.plot(kind='scatter', x='SkinThickness', y='Insulin', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
X = dataSet_diabetes.drop(columns = 'Outcome', axis=1)
Y = dataSet_diabetes['Outcome']
```

```python
X.head()
```

```python
ss = StandardScaler()
```

```python
scaled_X = ss.fit_transform(X)
```

```python
print(scaled_X)
```

```python
X = scaled_X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, stratify=Y, random_state=1)
```

```python
svc = svm.SVC(kernel='linear')
```

```python
svc.fit(X_train, Y_train)
```

```python
Y_pred = svc.predict(X_test)
```

```python
model_accuracy = accuracy_score(Y_pred, Y_test)
```

```python
print(model_accuracy)
```
